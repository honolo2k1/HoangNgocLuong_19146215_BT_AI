```python
import numpy as np
from numpy import cos
from numpy import sin
from math import radians
import csv
import math as m
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.utils.np_utils import to_categorical
from keras.models import Sequential
from sklearn.preprocessing import StandardScaler
from keras.layers import Dense, Activation, Dropout, Input
from tensorflow.keras.optimizers import RMSprop
from keras.callbacks import EarlyStopping
```

```python
l1 = 50
l2 = 40
l3 = 20
with open('robot_3bac.csv','w') as file:
  writer = csv.writer(file)
  writer.writerow(['theta1','theta2','theta3','phi','px','py'])
  for theta1 in range(-90,120,1):
    for theta2 in range(-90,120,1):
      for theta3 in range(-90,120,1):
        theta1 = (theta1*m.pi)/180
        theta2 = (theta2*m.pi)/180
        theta3 = (theta3*m.pi)/180
        px = l1*m.cos(theta1) + l2*m.cos(theta1+theta2)+l3*m.cos(theta1+theta2+theta3)
        py = l1*m.sin(theta1) + l2*m.sin(theta1+theta2)+l3*m.sin(theta1+theta2+theta3)
        phi = theta1 + theta2 + theta2
        writer.writerow([theta1,theta2,theta3,phi,px,py])
```

```python
data = pd.read_csv('robot_3bac.csv')
y = data.drop(data.columns[3:],axis = 1)
x = data.drop(data.columns[:3],axis = 1)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
model = Sequential()
model.add(Dense(512,activation='relu',input_shape=(3,)))
model.add(Dropout(0.2))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(3,activation='softmax'))
model.summary()
```

```python
model.compile(loss = 'mse', optimizer = RMSprop(), metrics = ['accuracy'])
history = model.fit(x_train,y_train,batch_size = 512,epochs = 10,verbose=1,callbacks=[EarlyStopping(monitor='val_loss',patience= 100)],validation_data = (x_test, y_test))
```

```python
score = model.evaluate(x_test,y_test,verbose=0)
print('Sai số: ',score[0])
print('Độ chính xác: ',score[1])
```
