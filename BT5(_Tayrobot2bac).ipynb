# TAY ROBOT 2 BAC

```python
import csv
import math as m
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Flatten
from tensorflow.keras.optimizers import RMSprop
from keras.callbacks import EarlyStopping

L1 = 50
L2 = 40
with open('robot_2bac.csv','w') as file:
  writer = csv.writer(file)
  writer.writerow(['theta1','theta2','px','py'])
  for theta1 in range(-180,180,1):
    for theta2 in range(-180,180,1):
      theta1 = (theta1*m.pi)/180
      theta2 = (theta2*m.pi)/180
      px = L1*m.cos(theta1) + L2*m.cos(theta1+theta2)
      py = L1*m.sin(theta1) + L2*m.sin(theta1+theta2)
      writer.writerow([theta1,theta2,px,py])
```

```python
data = pd.read_csv('robot_2bac.csv')
y = data.drop(data.columns[2:],axis = 1)
x = data.drop(data.columns[:2],axis = 1)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
model = Sequential()
model.add(Dense(128,activation='relu',input_shape=(2,)))
model.add(Dropout(0.2))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(2,activation='Softmax'))
print(model.summary())
```

```python
model.compile(loss='mse',optimizer=RMSprop(),metrics=['accuracy'])
history = model.fit(x_train,y_train,batch_size = 128,epochs = 100,verbose=1,callbacks=[EarlyStopping(monitor='val_loss',patience=20)],validation_data = (x_test, y_test))
```

```python
score = model.evaluate(x_test,y_test,verbose=0)
print('Sai số: ',score[0])
print('Độ chính xác: ',score[1])
```
